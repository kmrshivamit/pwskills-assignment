1. Supervised Learning:
- Uses labeled data.
- Goal: Make accurate predictions based on known inputs and outputs.
- Example: Image classification with labeled images.

2. Unsupervised Learning:
- Uses unlabeled data.
- Goal: Find patterns and structures without labeled outputs.
- Example: Customer segmentation based on behavior.

3. Semi-Supervised Learning:
- Uses a mix of labeled and unlabeled data.
- Goal: Improve model performance with limited labeled data.
- Example: Sentiment analysis using both labeled and unlabeled reviews.

Train Data: It is used for training the model so that it can predict it.

Test Data:  It is used for testing model accurracy.

Validation :It is used to validate the model and then we try to do hyper parameter tuning to improve accurracy


1. Training Data:
- Used for training the machine learning model.
- It consists of input features and corresponding labeled outputs.
- The model learns from this data during the training process to make accurate predictions.

2. Test Data:
- Used to evaluate the model's performance and accuracy.
- It is separate from the training data and contains input features but not the corresponding labels.
- The model makes predictions on the test data, and its performance is measured by comparing the predicted outputs with the true labels.

3. Validation Data:
- Used for hyperparameter tuning and model selection.
- It is a subset of the training data and helps assess the model's performance on data it hasn't seen during training.
- The model's hyperparameters (e.g., learning rate, number of hidden layers) can be adjusted based on the validation performance to improve accuracy.

By using separate training, validation, and test datasets, machine learning practitioners can build more robust and reliable models, avoiding overfitting to the training data and ensuring generalization to new, unseen data.

Unsupervised learning in anomaly detection learns normal patterns from data without labeled anomalies. 
It identifies anomalies as data points deviating significantly from the learned patterns, making it useful for 
scenarios with scarce labeled anomaly data.
