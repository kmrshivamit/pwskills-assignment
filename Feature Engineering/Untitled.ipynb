```python
# Skewing Descriptive Statistics: Outliers can greatly affect measures like the mean and standard deviation. 
# The mean is sensitive to extreme values, while the median is more robust to outliers.

# Inaccurate Models: Outliers can have a significant impact on regression and predictive models. Models might try to fit to 
# these extreme values, leading to poor generalization.

# Misinterpretation of Trends: Outliers can distort the interpretation of trends in your data, leading to incorrect 
# conclusions about relationships between variables.

# Biased Conclusions: If not handled properly, outliers might lead to biased or misleading conclusions about the underlying 
# population.
```

```python
df['Date_of_Journey']=pd.to_datetime(df['Date_of_Journey'])
```

```python

# Extract month, day, and weekday
df['Month'] = df['Date_of_Journey'].dt.month
df['Day'] = df['Date_of_Journey'].dt.day
df['Weekday'] = df['Date_of_Journey'].dt.weekday

# Plotting monthly trends
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Month')
plt.title('Monthly Flight Booking Trends')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.show()

# Plotting weekday analysis
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Weekday')
plt.title('Weekday Flight Booking Trends')
plt.xlabel('Weekday')
plt.ylabel('Number of Bookings')
plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```
