```python
# Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of
# features. What is the minimum number of principal components required to explain 90% of the variance in
# the data?
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load the wine quality dataset
url = "wineQualityReds.csv"
df = pd.read_csv(url)
df=df.iloc[:,1:]
```

```python
df.head()
```

```python
# Select only the numerical columns for PCA
numerical_columns = df.columns[:-1]  # Exclude the 'quality' column
data_for_pca = df[numerical_columns]
```

```python
# Standardize the data using StandardScaler
scaler = StandardScaler()
scaled_data_for_pca = scaler.fit_transform(data_for_pca)
```

```python
# Perform PCA
pca = PCA()
pca.fit(scaled_data_for_pca)

# Calculate the cumulative explained variance
cumulative_explained_variance = pca.explained_variance_ratio_.cumsum()

# Determine the minimum number of components for 90% variance explained
min_components_for_90_var = (cumulative_explained_variance >= 0.9).argmax() + 1

# Print the results
print(f"Explained variance by each component: {pca.explained_variance_ratio_}")
print(f"Cumulative explained variance: {cumulative_explained_variance}")
print(f"Minimum number of components to explain 90% variance: {min_components_for_90_var}")

# Transform the data using the selected number of components
pca = PCA(n_components=min_components_for_90_var)
data_pca = pca.fit_transform(scaled_data_for_pca)

# Display the first few rows of the transformed data
df_pca = pd.DataFrame(data_pca, columns=[f'PC{i+1}' for i in range(min_components_for_90_var)])
df_pca['quality'] = df['quality']  # Add back the 'quality' column
df_pca.head()
```
