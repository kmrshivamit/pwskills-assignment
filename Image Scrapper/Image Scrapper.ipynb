```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url='https://www.youtube.com/@PW-Foundation/videos'
```

```python
html_response=requests.get(url)
```

```python
html_response.text
```

```python
html_parsed=bs(html_response.text,'html.parser')
```

```python
html_parsed
```

```python
video_box=html_parsed.find_all('div')
```
