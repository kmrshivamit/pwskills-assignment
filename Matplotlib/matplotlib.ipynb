```python
# Line plot: A line plot is a basic plot used to display the trend of a variable over time. 
#     It can be created using the plot() function.

# Scatter plot: A scatter plot is used to visualize the relationship between two variables. 
#     It can be created using the scatter() function.

# Bar plot: A bar plot is used to compare the values of different categories.
#     It can be created using the bar() or barh() function.

# Histogram: A histogram is used to display the distribution of a continuous variable. 
#     It can be created using the hist() function.

# Pie chart: A pie chart is used to display the proportion of each category in a dataset.
#     It can be created using the pie() function.
```

# Line Plot

```python
from matplotlib import pyplot as plt
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the data for the four line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots arranged in a 2 x 2 grid
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

# Plot the four line plots in each of the subplots
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Set the titles and labels for the subplots
axs[0, 0].set_title('Line Plot 1')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')

axs[0, 1].set_title('Line Plot 2')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')

axs[1, 0].set_title('Line Plot 3')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')

axs[1, 1].set_title('Line Plot 4')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')

# Adjust the spacing between the subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a list
data = [box1, box2]

# Create a box plot
fig, ax = plt.subplots()
ax.boxplot(data)

# Add labels
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')

# Show the plot
plt.show()
```
